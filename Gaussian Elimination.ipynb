# Gaussian Elimination #

### Summary of Advantage ###  

1. 利用了 SymPy 的符号引擎，能够以精确的分数形式进行运算，消除了浮点数舍入误差。  
Utilizes SymPy’s symbolic engine to perform calculations with exact fractions, eliminating floating-point rounding errors.

2. 通过调用 rref() 函数，代码自动执行了复杂的“行最简形矩阵”变换。  
By calling the rref() function, the code automatically executes complex Row Reduced Echelon Form transformations.

3. 代码使用 f-string 根据矩阵维度动态生成变量名。  
The code uses f-strings to dynamically generate variable names based on matrix dimensions.

4. 结合 IPython 的渲染功能，代码能以标准的 LaTeX 格式输出矩阵。  
Combined with IPython’s rendering capabilities, the code outputs matrices in standard LaTeX format.

### Example ###

```python
import sympy as sp
from IPython.display import display, Math

# 1. 定义增广矩阵 Define the Augmented Matrix 
matrix_data = [
    [2, -1,  0,  0],
    [-1, 2, -1, -1],
    [0, -3,  4,  4]
]

# 2. 获取行数 Extract Number of Equatoins
num_vars = len(matrix_data) 

# 3. 生成n个变量 Generate Variables from x1 to x_n
vars = sp.symbols(f'x1:{num_vars + 1}') 

# 4. 构造增广矩阵 Construct the Augmented Matrix
M = sp.Matrix(matrix_data)

print("Original Augmented Matrix [A|b]:")
display(M)

# 5. 执行高斯-约旦消元法 Perform Gauss-Jordan Elimination
rref_matrix, pivots = M.rref()

print("\nRow Reduced Echelon Form (RREF):")
display(rref_matrix)

# 5. 提取结果 Extract the solutions
solutions = {vars[i]: rref_matrix[i, 3] for i in range(len(vars))}

print("\nSolution:")
for var, val in solutions.items():
    print(f"{var} = {val}")
```
